## 0.2.b

$P_r = P_t + 20 \cdot \log_{10}(\frac{c}{4 \cdot \pi \cdot f \cdot r})$

Move everything not in the log to the left:

$\frac{P_r - P_t}{20}= \log_{10}(\frac{c}{4 \cdot \pi \cdot f \cdot r})$

remove the log on the right side by raising to the power of 10:

$10^{\frac{P_r - P_t}{20}}= \frac{c}{4 \cdot \pi \cdot f \cdot r}$

to the power -1:

$\frac{1}{10^{\frac{P_r - P_t}{20}}}= \frac{4 \cdot \pi \cdot f \cdot r}{c}$

solve:

$r = \frac{c}{10^{\frac{P_r - P_t}{20}} \cdot 4 \cdot \pi \cdot f} $





## 0.2.c

difference in distance r, between signal strengths of -30 and 31 and -60 and -61.

```python
c = 299792458
f = 2.4e9
Pt = 0

# function to calculate the radius (distance)
def radius(c, f, Pt, Pr):
    return (c/((10**((Pr-Pt)/20.0))*4*np.pi * f))


print 'Difference in radius between a signal strength of -30 and -31: %0.3f'% (radius(c,f,Pt,-31) - radius(c,f,Pt,-30))
print 'Difference in radius between a signal strength of -60 and -61: %0.3f'% (radius(c,f,Pt,-61) - radius(c,f,Pt,-60))
```

## Answer 0.2 c & d:

As can be seen the difference in radius becomes larger when the signal strength is weaker. Thus uncertainty becomes larger as the signal strength decreases. The mathematical explanation for this is because in the following formula for the radius:

$r = \frac{c}{10^{\frac{P_r - P_t}{20}} \cdot 4 \cdot \pi \cdot f} $

as P_r becomes smaller, the exponent becomes more negative, this causes the denominator to become smaller and the outcome to be much larger in total. Hence small perturbations will cause more of a change.

## Answer 0.3.f

Friis Equation:
$P_r = P_t + 20 \cdot \log_{10}(\frac{c}{4 \cdot \pi \cdot f \cdot r}) $

We know that r is simply the euclidean distance between the phone and router in 3d space. r can thus be expressed as follows:

$r = \sqrt{((x-x_i)^2+(y-y_i)^2+(z-z_i)^2)}$

What we can do is we can separate the fraction inside the logarithm firstly, and separate the logs:

$P_r = P_t + 20 \cdot \log_{10}(\frac{c}{4 \cdot \pi \cdot f }\cdot \frac{1}{r})$

$P_r = P_t + 20 \cdot \log_{10}(\frac{c}{4 \cdot \pi \cdot f }) - 20 \log_{10}(r)$

Substitute for r:

$P_r = P_t + 20 \cdot \log_{10}(\frac{c}{4 \cdot \pi \cdot f }) - 20 \log_{10}(((x-x_i)^2+(y-y_i)^2+(z-z_i)^2)^{\frac{1}{2}})$

moving the exponent outside of the log:

$P_r = P_t + 20 \cdot \log_{10}(\frac{c}{4 \cdot \pi \cdot f }) - 10 \log_{10}((x-x_i)^2+(y-y_i)^2+(z-z_i)^2)$

keeping $(z-z_i)$ constant, this yields:

$P^i_r(x,y) = P_t + 20 \cdot \log_{10}(\frac{c}{4 \cdot \pi \cdot f }) - 10 \log_{10}((x-x_i)^2+(y-y_i)^2+Z^2)$

```python
# coordinates of the router and the mobile
coords_r = [0,0]
coords_m = [20, 0]

c = 299792458
f = 2.4
Pt = 0

def Friis_equation(radius, factor):
    return P_t + factor* np.log10(c/(4*np.pi*f*radius))

# calculate the radius with z^2 fixed at 2
def rad(r, m):
    return np.sqrt((r[0]-m[0])**2+(r[1]-m[1])**2 + 4)

# Basically the Friis formula with explicit r calculation
def Friis(r, m, sigma, num = 1000, factor = 20):
        
    gaussian = np.random.normal(0,sigma, num)
    
    # calculating the radius
    radius = np.asarray([rad(r,m) for x in range(num) ])
    
    # calculate value friis equation
    P_r = Friis_equation(radius, 20)
    P_r_dist = Friis_equation(radius, factor) + gaussian
    
    # return the residuals
    return P_r_dist - P_r
    
    
print "Experiment 1 (0.3.g) "
exp_1 = Friis(coords_r, coords_m, 1)



plt.hist(exp_1, 50, normed=True)

plt.show()

print 'Mean (Expected value) of pull distribution exp 1: %0.4f'% (np.mean(exp_1))
print 'st. dev of pull distribution exp 1: %0.4f'% (np.std(exp_1))
```

## Answer 0.3.j 

Firstly in your pull distribution you want to see that you have a mean of 0. This means that the average of your residuals is 0, otherwise this is indication of measurement/systematic error in your model and you would have to correct this somehow.
```python
coords_wifi = [[0,0],[0,20],[20,20],[20,0]]
coords_phone = np.array([[x,5] for x in range(-5,15)])

def Friis_equation(radius, factor, P_t = 0, c = 299792458, f = 2.4):
    return P_t + factor* np.log10(c/(4*np.pi*f*radius))

# calculate the radius with z^2 fixed at 2
def rad(r, m):
    return np.sqrt((r[0]-m[0])**2+(r[1]-m[1])**2 + 4)

# Basically the Friis formula with explicit r calculation
def Friis(r, m, sigma, num = 1, factor = 20, rand = False):
    
    # calculating the radius
    radius = np.asarray([rad(r,m) for x in range(num) ])
    
    # calculate value friis equation
    P_r = Friis_equation(radius, factor)
    
    if rand:
        P_r += np.random.normal(0,sigma, num)
        
        
    return P_r

# Function for calculating the chisquare
def ChiSquare(S,P):
    return sum(((S-P)**2)/1)

# calculate the true signal
trueSignal = []
for coord in coords_wifi:
    trueSignal.append(Friis(coord,[5,5], 1))

trueSignal = np.array(trueSignal)

# calculate the estimates for the different coordinates
estimates = []
for m in coords_phone:
    temp = []
    for r in coords_wifi:
        temp.append(Friis(r, m, 4, rand = False))
    
    estimates.append(np.array(temp))

# calculate the chi2
chi2 = []
for x in estimates:
    chi2.append(ChiSquare(trueSignal, x))

# plot the chi2
plt.plot( [x[0] for x in coords_phone ],chi2)
plt.ylabel('Chi2 at receiver')
plt.xlabel('x coordinate phone')
plt.show()
    
```

```python
def minimums(coords_phone, coords_wifi, trueSignal):
    
    # calculate the estimates for the different coordinates
    estimates = []
    for m in coords_phone:
        temp = []
        for r in coords_wifi:
            temp.append(Friis(r, m, 1, rand = True))

        estimates.append(np.array(temp))

    # calculate the chi2
    chi2 = []
    for x in estimates:
        chi2.append(ChiSquare(trueSignal, x))

    return chi2

# calculate the chi2
chi2 = minimums(coords_phone, coords_wifi, trueSignal)
    
# plot the chi2
plt.plot( [x[0] for x in coords_phone ],chi2)
plt.ylabel('Chi2 at receiver')
plt.xlabel('x coordinate phone')
plt.show()

print "If this is repeated several times, the minimum changes position."
```

```python
def Friis_equation(radius, factor, P_t = 0, c = 299792458, f = 2.4):
    return P_t + factor* np.log10(c/(4*np.pi*f*radius))

# calculate the radius with z^2 fixed at 2
def rad(r, m):
    return np.sqrt((r[0]-m[0])**2+(r[1]-m[1])**2 + 4)

# Basically the Friis formula with explicit r calculation
def Friis(r, m, sigma, factor = 20):
    
    # calculating the radius
    radius = np.asarray(rad(r,m))
    
    # calculate value friis equation
    P_r = Friis_equation(radius, factor) 
        
    return P_r

# Function for calculating the chisquare
def ChiSquare(S,P):
    return np.sum((((S-P))/1)**2)

def to_optimize(x, trueSignal, coords_wifi, noise, sigma):
    estimates = []
    
    for num, y in enumerate(coords_wifi):
        estimates.append(Friis(y, x, 1) + noise[num])
    
    return ChiSquare(trueSignal, np.array(estimates))



def simulate_packets(sigma):
    mins =[]
    coords=[]
    for x in range(1000):
        noise = list(np.random.normal(0,sigma,4))
        temp = opt.minimize(to_optimize, x0 =[0,0], args=(trueSignal, coords_wifi, noise, sigma))
        coords.append(tuple(temp.x))
        mins.append(temp['fun'])
    return mins, coords

mins, coords = simulate_packets(1)

x,y = zip(*coords)[0], zip(*coords)[1]
optimum = np.mean(mins)


print 'The coordinates of the minimum are for x and y respectively %f0 and %f0. The minimum\'s average value is: %f0' % (np.mean(x), np.mean(y), optimum)
```

### Question 0.5 s
Substitute the Friis approximation in the chi square equation:

$\chi^2 = \sum\frac{(S_i - P^r_i(x,y))^2}{\sigma^2_i} $

Now we substitute :

$\chi^2 = \sum\frac{(S_i - (P^r_i(x_0,y_0)+ \alpha (x-x_0) + \beta (y-y_0)))^2}{\sigma^2_i} $ 

where, $\alpha = \frac{\partial P^r_i(x_0,y_0)}{\partial x}$ and $\beta = \frac{\partial P^r_i(x_0,y_0)}{\partial y}$

Now we expand this and collect terms, we obtain:

$\chi^2 = \sum \frac{((S_i - P^r_i(x_0,y_0))^2 + (\alpha(x - x_0)+\beta(y-y_0))^2 - 2(S_i - P^r_i(x_0,y_0))\alpha(x-x_0)-2(S_i - P^r_i(x_0,y_0))\beta(y-y_0)}{\sigma_i}$

### Question 0.5 V
How large does $\Delta x $ have to be to raise the $chi^2$ by 1? (same question for $\Delta x $)

We know the following for $\Delta x$:

$\Delta x^2 \sum \alpha_i + 2 \Delta x \Delta y \sum \alpha_i \beta_i + \Delta y^2 \sum \beta^2 -1 = 0 $

We can say the following:

a= $\sum \alpha_i^2$, b = $2 \Delta y \sum \alpha_i \beta_i$, c = $\Delta y^2 \sum \beta^2 -1 $ 

Determine the value for the discriminant:

D = $\sqrt{(2 \Delta y \sum \alpha_i \beta_i)^2 - 4\sum \alpha_i^2 (\Delta y^2 \sum \beta^2 - 1)}$

Simplify:

D = $2\sqrt{ \Delta y^2 (\sum \alpha_i \beta_i)^2 - \sum \alpha_i \Delta y^2 \sum \beta^2 +\sum \alpha_i^2 }$

D = $2\sqrt{ \Delta y^2 ((\sum \alpha_i \beta_i)^2 - \sum \alpha_i \sum \beta^2 )+ \sum \alpha_i^2 }$

Now we can solve for $\Delta y$:


$\Delta y^2 ((\sum \alpha_i \beta_i)^2 - \sum \alpha_i \sum \beta^2 )+ \sum \alpha_i^2  = 0$

$\Delta y^2  = \frac{\sum \alpha_i^2}{( \sum \alpha_i \sum \beta^2- (\sum \alpha_i \beta_i)^2 )} $

If we wish to solve for y we turn this procedure around and we get the following answer for the discriminant:

$\Delta x^2  = \frac{\sum \beta_i^2}{( \sum \alpha_i \sum \beta^2- (\sum \alpha_i \beta_i)^2 )}$

if we substitute these values into the for x and y into the discriminant we get 0. At these locations we thus find that there is only 1 optimal solution.

Now we can easily write down the optimal values for which the chi square increases by 1 for the $\Delta x$ and $\Delta y$ values.

These are:

$\Delta \chi^2 = -\frac{\sum \alpha_i \beta_i}{\sum \alpha_i^2}\Delta y$

$\Delta \chi^2 = -\frac{\sum \alpha_i \beta_i}{\sum \beta_i^2}\Delta x$



```python
import math as mt
from matplotlib.patches import Ellipse

# derivative
def der(x,r):
    return (-20*(x[0]-x[1])/(r*np.log(10)))

# calculate the radius with z^2 fixed at 2
def denom(r, m):
    return ((r[0]-m[0])**2+(r[1]-m[1])**2 + 4)

# Function for calculating the chisquare
def ChiSquare(S,P):
    return np.sum((((S-P))/1)**2)

def to_optimize(x, trueSignal, coords_wifi, noise, sigma):
    estimates = []
    
    for num, y in enumerate(coords_wifi):
        estimates.append(Friis(y, x, 1) + noise[num])
    
    return ChiSquare(trueSignal, np.array(estimates))

def calc_uncertainty(x, coords_wifi, sigma = 1):
    
    # get the radiusses
    radii = [(denom(x,y), y[0],y[1]) for y in coords_wifi]
    
    # calculate the matrix elements
    inv_var_y = sum([der([x[1],r[2]], r[0])**2 for r in radii])/sigma
    inv_var_x = sum([der([x[0],r[1]], r[0])**2 for r in radii])/sigma
    inv_cov = sum([der([x[0],r[1]], r[0])* der([x[1],r[2]], r[0]) for r in radii])/sigma
    
    # get the inverse
    Binv = np.array([[inv_var_x, inv_cov ],[inv_cov,inv_var_y]])
    B = np.linalg.inv(Binv)
    
    std_y = np.sqrt((B[1][1])**-1)
    std_x = np.sqrt((B[0][0])**-1)
    
    
    return std_x, std_y

noise = list(np.random.normal(0,1,4))
temp = opt.minimize(to_optimize, x0 =[0,0], args=(trueSignal, coords_wifi, noise, 1))
x_std, y_std = calc_uncertainty(tuple(temp.x), coords_wifi)



print 'x,y coordinates of optimum: ', tuple(temp.x)
print 'std_x and std_y: ', x_std, y_std
plt.figure()
ax = plt.gca()

ellipse = Ellipse(xy=tuple(temp.x), width=x_std, height=y_std, 
                        edgecolor='r', fc='None', lw=2)
ax.add_patch(ellipse)
plt.scatter(temp.x[0], temp.x[1])
plt.axis([0,20,0,20])
plt.show()

```

```python
# Dictionary with the wifi locations
wifi_loc = {'Lima': [5.82, 5.48], 'Mike': [11.33, 9.43], 'Kilo': [12.39, 6.77],
            'Oscar': [2.48, 7.36], 'Alpha':[8.53, 2.16], 'India': [2.18, 5.61],
            'November':[8.34, 4.13],'Hotel':[5.43, 4.71], 'Romeo': [10.99, 5.94],
            'Quebec':[6.82, 9.78], 'Papa': [9.9, 10.39]}

wx = zip(*[x for x in wifi_loc.values()])[0]
wy = zip(*[x for x in wifi_loc.values()])[1]


# previous work, reused
def Friis_equation(radius, Pt , c = 299792458.0, f = 2.4e9):
    return Pt + 20.0*np.log10(c/(4.0*np.pi*f*radius)) 

# calculate the radius with z^2 fixed at 2
def rad(x, y):
    return np.sqrt(np.power(y[0]-x[0], 2)+np.power(y[1]- x[1],2) + 4.0)

# Basically the Friis formula with explicit r calculation
def Friis(y, x, Pt):
    
    # calculating the radius
    radius = np.asarray(rad(x,y))
    
    # calculate value friis equation
    # include the radius as the third element of x
    P_r = Friis_equation(radius, Pt) 
        
    return P_r

# Function for calculating the chisquare
def ChiSquare(S,P, sigma):
    return np.sum(((np.power(S-P,2))/(sigma**2)))

def to_optimize(x, trueSignal, coords_wifi,sigma):

    estimates = []
    coords = [x[0],x[1]]
    Pt = x[2]
    
    for num, y in enumerate(coords_wifi):
        estimates.append(Friis(y, coords, Pt))
    
    return ChiSquare(trueSignal, np.array(estimates), sigma)


chi = []
vals = []
stds = []
def get_locations(group):
    sigma = 6
    
    # get the signal strengths and locations of the wifi drones
    wifi_data = []
    
    # if it is shorter than 2 don't do anything!
    if len(group['droneId'].values) < 4:
        return group
    
    # iterate over the wifi data
    for code in group['droneId'].values:
        wifi_data.append((wifi_loc[code], float(group[group['droneId'] == code].values.tolist()[0][7])))
    
    coords = zip(*wifi_data)[0]
    trueSignal = zip(*wifi_data)[1]
    
    temp = opt.minimize(to_optimize, x0 =[0,0,0], args=(trueSignal, coords, sigma), method="L-BFGS-B")
    
    # store the chi2 values and optimal x,y coordinates
    chi.append(temp['fun'])
    vals.append(temp['x'])
    
    # Calculate the uncertainties
    stds.append(tuple(calc_uncertainty(temp['x'], coords, sigma)))
    
    return group
    


df = df.groupby(['mean'] ).apply(get_locations)
x = zip(*vals)[0]
y = zip(*vals)[1]

print 'Path through the Restaurant, red dots indicate wifi locations and blue dots positions'
plt.scatter(wx, wy, color = 'r')
plt.plot(x,y, '-o')
plt.show()

```
